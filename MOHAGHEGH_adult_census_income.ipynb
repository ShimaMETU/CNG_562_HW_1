```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib import pyplot as plt
from sklearn import preprocessing
adult_census_income = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.data', 
                                  names = ["Age", "Workclass","fnlwgt","education", "education-num","marital-status",
                                           "occupation","relationship","race", "sex", "capital.gain","capital.los",
                                           "hours.per.week","native.country", "income"] , na_values=["?"])
adult_census_income.info()
```

```python
adult_census_income.describe()
```

```python
#checking missing valus since in dataset I saw that the unknown values shown via ? I replace with null
adult_census_income.replace('[?]', np.nan, regex=True , inplace=True)
adult_census_income
```

```python
adult_census_income.isnull().sum()
```

```python
adult_census_income['Workclass'].describe()
```

```python

adult_census_income['Workclass'].replace( np.nan, 'Private' ,regex=True , inplace=True)
```

```python
adult_census_income['occupation'].describe()
```

```python

adult_census_income['occupation'].replace( np.nan, 'Prof-specialty' ,regex=True , inplace=True)
```

```python

adult_census_income['native.country'].describe()
```

```python
adult_census_income['native.country'].replace( np.nan, 'United-States' ,regex=True , inplace=True)
```

```python

adult_census_income.isnull().sum()
```

```python
from sklearn.preprocessing import MinMaxScaler,StandardScaler

data = adult_census_income
target = data['income']
del data['income']
del data['education']
print(adult_census_income)
data = pd.get_dummies(data, prefix=['Workclass','marital-status','occupation','relationship','race','sex','native.country'], columns=['Workclass','marital-status','occupation','relationship','race','sex','native.country'])
scaler = MinMaxScaler()
data[['Age','fnlwgt','education-num','capital.gain','capital.los', 'hours.per.week']] = scaler.fit_transform(data[['Age','fnlwgt','education-num','capital.gain','capital.los', 'hours.per.week']])

print()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(data, target, train_size = 0.7)
X_train.shape, X_test.shape, Y_train.shape, Y_test.shape
```

```python
############################################### My Logistic Regression#################################
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression(random_state=0)
lr = lr.fit(X_train,Y_train)
predicted_lr = lr.predict(X_test)
from sklearn.metrics import accuracy_score
from sklearn import metrics
metrics.accuracy_score(Y_test, predicted_lr)
```

```python
#################################### Cross validation Logistic Regression ####################################################
from sklearn.model_selection import cross_val_score  
all_accuracies = cross_val_score(estimator=lr, X=X_train, y=Y_train, cv=5)
all_accuracies.mean()
```

```python
############################################ Grid search  for logistic Regression #######################################################
from sklearn.model_selection import GridSearchCV
param_grid = {'C' :np.arange(0.1,1)}
reg = LogisticRegression()
grid = GridSearchCV(lr, param_grid, cv = 5)
grid.fit(X_train, Y_train)
grid.best_params_
```

```python
############################################ My Decision tree#####################################################3
from sklearn.tree import DecisionTreeClassifier
dtc = DecisionTreeClassifier(random_state=1, max_depth=5)
dtc = dtc.fit(X_train, Y_train)
predicted_dtc = dtc.predict(X_test)
from sklearn import metrics
metrics.accuracy_score(Y_test, predicted_dtc)
```

```python
#################################### Cross validation Decision tree ####################################################
from sklearn.model_selection import cross_val_score  
all_accuracies = cross_val_score(estimator=dtc, X=X_train, y=Y_train, cv=5)
all_accuracies.mean()
```

```python
############################################ My Grid search for Decision tree #######################################################
#### !!!!!!!!!!!this one takes a lot of time ten mintues!!!!!!!!!!!!!!!############
from sklearn.model_selection import GridSearchCV
param_grid = {'max_depth': range(2,15),'max_features': range(4,93)}
grid = GridSearchCV(dtc, param_grid, cv=5, n_jobs=-1, verbose=True)

grid.fit(X_train, Y_train)
y_pred_class = grid.predict(X_test)
grid.best_params_
```

```python
############################################## My SVC ###############################################################
from sklearn.svm import SVC
svc = SVC(kernel='rbf', C=1.0, degree=3, class_weight=None)
svc =svc.fit(X_train, Y_train)
predicted_svc = svc.predict(X_test)
from sklearn import metrics
print (metrics.accuracy_score(Y_test, predicted_svc))

import time
import datetime
start_time = time.time()
print (f"Elapsed time: {datetime.timedelta(seconds=time.time()-start_time)} ")
```

```python
#################################### My Cross  for validation SVC ####################################################
from sklearn.model_selection import cross_val_score  
all_accuracies = cross_val_score(estimator=svc, X=X_train, y=Y_train, cv=5)
all_accuracies.mean()
```

```python
############################################ My Grid search for SVC #######################################################
#### !!!!!!!!!!!this one takes a lot of time!!!!!!!!!!!!!!!############
from sklearn.model_selection import GridSearchCV
param_grid = {'C' :np.arange(0.1,1)}
svc = SVC()
grid = GridSearchCV(svc, param_grid, cv=3)
grid.fit(X_train, Y_train)
grid.best_params_
```

```python
################################################### My KNN #############################################
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=6,metric='minkowski',p=2)
knn =knn.fit(X_train, Y_train)
predicted_knn = knn.predict(X_test)
from sklearn import metrics
metrics.accuracy_score(Y_test, predicted_knn)
```

```python
#################################### My Cross validation KNN ####################################################
from sklearn.model_selection import cross_val_score  
all_accuracies = cross_val_score(estimator=knn, X=X_train, y=Y_train, cv=5)
all_accuracies.mean()
```

```python
############################################ My Grid search KNN #######################################################
#### !!!!!!!!!!!this one takes a lot of time!!!!!!!!!!!!!!!############
from sklearn.model_selection import GridSearchCV
param_grid = {'n_neighbors' :np.arange(1,10)}
knn = KNeighborsClassifier()
grid = GridSearchCV(knn, param_grid, cv = 5)
grid.fit(X_train, Y_train)
grid.best_params_
```
