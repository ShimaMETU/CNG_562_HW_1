```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib import pyplot as plt
from sklearn import preprocessing
MOHAGHEGH_adult_census_income = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.data', 
                                  names = ["Age", "Workclass","fnlwgt","education", "education-num","marital-status",
                                           "occupation","relationship","race", "sex", "capital.gain","capital.los",
                                           "hours.per.week","native.country", "income"] , na_values=["?"])
MOHAGHEGH_adult_census_income
```

```python
MOHAGHEGH_adult_census_income.info()
```

```python
MOHAGHEGH_adult_census_income.describe()
```

```python
#checking missing valus since in dataset I saw that the unknown values shown via ? I replace with null
MOHAGHEGH_adult_census_income.replace('[?]', np.nan, regex=True , inplace=True)
MOHAGHEGH_adult_census_income
```

```python
MOHAGHEGH_adult_census_income.isnull().sum()
```

```python
#Since we have empty cells at 'Workclass' ,'occupation' and 'native.country ' I replace them with most frequent ones
MOHAGHEGH_adult_census_income['Workclass'].describe()
```

```python
MOHAGHEGH_adult_census_income['Workclass'].replace( np.nan, 'Private' ,regex=True , inplace=True)
```

```python
MOHAGHEGH_adult_census_income['occupation'].describe()
```

```python
MOHAGHEGH_adult_census_income['occupation'].replace( np.nan, 'Prof-specialty' ,regex=True , inplace=True)
```

```python
MOHAGHEGH_adult_census_income['native.country'].describe()
```

```python
MOHAGHEGH_adult_census_income['native.country'].replace( np.nan, 'United-States' ,regex=True , inplace=True)
```

```python
MOHAGHEGH_adult_census_income.isnull().sum()
```

```python
###############################################preparing dataset for ML ######################################################
MOHAGHEGH_adult_census_income['income'].replace('<=50K','0',regex=True, inplace = True)
MOHAGHEGH_adult_census_income['income'].replace('>50K','1',regex=True,inplace = True)
#in Workclass colums to reduce variables
self_employed = ['Self-emp-not-inc','Self-emp-inc']
govt_employees = ['Local-gov','State-gov','Federal-gov']
MOHAGHEGH_adult_census_income['Workclass'].replace(to_replace = self_employed , value = 'Self_employed',regex=True,inplace = True)
MOHAGHEGH_adult_census_income['Workclass'].replace(to_replace = govt_employees, value = 'Govt_employees',regex=True,inplace = True)
#in education colum to reduce variables
hs_grad = ['HS-grad','11th','10th','9th','12th']
elementary = ['1st-4th','5th-6th','7th-8th']
MOHAGHEGH_adult_census_income['education'].replace(to_replace = hs_grad,value = 'HS-grad',regex=True, inplace = True)
MOHAGHEGH_adult_census_income['education'].replace(to_replace = elementary,value = 'elementary_school',regex=True,inplace = True)
#in marital.status colum to reduce variable
married= ['Married-spouse-absent','Married-civ-spouse','Married-AF-spouse']
separated = ['Separated','Divorced']
MOHAGHEGH_adult_census_income['marital-status'].replace(to_replace = married , value = 'Married',regex=True, inplace = True)
MOHAGHEGH_adult_census_income['marital-status'].replace(to_replace = separated,value = 'Separated',regex=True, inplace = True)
#since some of the colums is kind of dublicate or extra data I can be deleted
#Education num and education are giving similar information also Relationship and marital status imply similar information.Hence keeping only one of the two
MOHAGHEGH_adult_census_income.drop('capital.gain' ,axis = 1,inplace = True)
MOHAGHEGH_adult_census_income.drop('capital.los' ,axis = 1,inplace = True)
MOHAGHEGH_adult_census_income.drop('education-num' , axis = 1, inplace = True)
MOHAGHEGH_adult_census_income.drop('relationship' , axis = 1, inplace = True)
MOHAGHEGH_adult_census_income.drop('fnlwgt' , axis = 1, inplace = True)
MOHAGHEGH_adult_census_income
```

```python
from sklearn.preprocessing import LabelEncoder
lb_make = LabelEncoder()
MOHAGHEGH_adult_census_income['Workclass'] = lb_make.fit_transform(MOHAGHEGH_adult_census_income['Workclass'])
MOHAGHEGH_adult_census_income['education'] = lb_make.fit_transform(MOHAGHEGH_adult_census_income['education'])
MOHAGHEGH_adult_census_income['marital-status'] = lb_make.fit_transform(MOHAGHEGH_adult_census_income['marital-status'])
MOHAGHEGH_adult_census_income['race'] = lb_make.fit_transform(MOHAGHEGH_adult_census_income['race'])
MOHAGHEGH_adult_census_income['sex'] = lb_make.fit_transform(MOHAGHEGH_adult_census_income['sex'])
MOHAGHEGH_adult_census_income['occupation'] = lb_make.fit_transform(MOHAGHEGH_adult_census_income['occupation'])
MOHAGHEGH_adult_census_income['native.country'] = lb_make.fit_transform(MOHAGHEGH_adult_census_income['native.country'])
MOHAGHEGH_adult_census_income
```

```python
from sklearn.utils import shuffle
adult_census_income = MOHAGHEGH_adult_census_income
adult_census_income = shuffle(adult_census_income)
target = adult_census_income['income']
del adult_census_income['income']
adult_census_income = pd.get_dummies(adult_census_income)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(adult_census_income, target, train_size = 0.7)
X_train.shape, X_test.shape, Y_train.shape, Y_test.shape
```

```python
############################################### My_Logistic_Regression #################################
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
lr = lr.fit(X_train,Y_train)
predicted_lr = lr.predict(X_test)
from sklearn.metrics import accuracy_score
from sklearn import metrics
metrics.accuracy_score(Y_test, predicted_lr)
```

```python
#################################### My_Cross_validation_for_Logistic _Regression ####################################################
from sklearn.model_selection import cross_val_score
all_accuracies = cross_val_score(estimator=lr, X=X_train, y=Y_train, cv=5)
all_accuracies.mean()
```

```python
############################################MY_Grid_search_logistic_Regression #######################################################
from sklearn.model_selection import GridSearchCV
param_grid = {'C' :np.arange(0.1,2)}
reg = LogisticRegression()
grid = GridSearchCV(lr, param_grid, cv = 5)
grid.fit(X_train, Y_train)
grid.best_params_
```

```python
############################################MY_Decision_Tree#################################################
from sklearn.tree import DecisionTreeClassifier
dtc = DecisionTreeClassifier(random_state=1000, max_depth=5)
dtc = dtc.fit(X_train, Y_train)
predicted_dtc = dtc.predict(X_test)
from sklearn import metrics
metrics.accuracy_score(Y_test, predicted_dtc)
```

```python
####################################MY_Cross_validation_for_Decision_Tree ####################################################
from sklearn.model_selection import cross_val_score
all_accuracies = cross_val_score(estimator=dtc, X=X_train, y=Y_train, cv=5)
all_accuracies.mean()
```

```python
############################################My_Grid_search_for_Decision_Tree#######################################################
from sklearn.model_selection import GridSearchCV
param_grid = {"criterion": ["gini", "entropy"], "min_samples_split": [2, 10], "max_depth": np.arange(2,10)}
grid = GridSearchCV(dtc, param_grid, cv=3, scoring='neg_mean_squared_error')
grid.fit(X_train, Y_train.values.ravel())
grid.best_params_
```

```python
##############################################MY_SVC###############################################################
from sklearn.svm import SVC
svc = SVC(kernel='rbf', C=1.0, degree=3, class_weight=None, gamma = 'auto' , max_iter=10)
svc =svc.fit(X_train, Y_train)
predicted_svc = svc.predict(X_test)
from sklearn import metrics
metrics.accuracy_score(Y_test, predicted_svc)
```

```python
####################################My_Cross_validation_for_SVC ####################################################
from sklearn.model_selection import cross_val_score  
all_accuracies = cross_val_score(estimator=svc, X=X_train, y=Y_train, cv=5)
all_accuracies.mean()
```

```python
############################################My_Grid_search-for_SVC #######################################################
from sklearn.model_selection import GridSearchCV
param_grid = {'C' :np.arange(0.1,1)}
svc = SVC()
grid = GridSearchCV(svc, param_grid, cv=3)
grid.fit(X_train, Y_train)
grid.best_params_
```

```python
###################################################My_KNN#############################################
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=6,metric='minkowski',p=2)
knn =knn.fit(X_train, Y_train)
predicted_knn = knn.predict(X_test)
from sklearn import metrics
metrics.accuracy_score(Y_test, predicted_knn)
```

```python
####################################Cross validation KNN ####################################################
from sklearn.model_selection import cross_val_score  
all_accuracies = cross_val_score(estimator=knn, X=X_train, y=Y_train, cv=5)
all_accuracies.mean()
```

```python
############################################Grid search KNN #######################################################
from sklearn.model_selection import GridSearchCV
param_grid = {'n_neighbors' :np.arange(1,10)}
knn = KNeighborsClassifier()
grid = GridSearchCV(knn, param_grid, cv = 5)
grid.fit(X_train, Y_train)
grid.best_params_
```
